# TCDF for COVID-19 Time Series Data Analysis

No we run TCDF on COVID-19 data sets for Germany, Italy and the US. 

## Germany

First, we use kernel size 7 and learning rate 0.001 with a significance of 0.8, training for 10,000 epochs. The nets converge nicely for many target variables with these settings and a graph-based visualization documents the TCDF-identified causal links intuitively. (You might need to scroll down in-line in order to see our visualization of the results!)

```python
%matplotlib inline
%run -i "runTCDF.py" --data data/covid-ger-extended-trim.csv --epochs=10000 --kernel_size=7 --hidden_layers=0 --learning_rate=0.001 --optimizer="Adam" --dilation_coefficient=7 --significance=0.8 --plot
```

Now, we increase the kernel size to 14 in order to include more context and also adapt the dilation coefficient accordingly. The resulting visualization shows significantly different causal links.

```python
%run -i "runTCDF.py" --data data/covid-ger-extended-trim.csv --epochs=10000 --kernel_size=14 --hidden_layers=0 --learning_rate=0.001 --optimizer="Adam" --dilation_coefficient=14 --significance=0.8 --plot
```

In another experiment, one hidden layer is introduced. The other hyper paramters are fixed as before. This leads to significantly longer lagging termporal cause links, as can be observed in the resulting graph.

```python
%run -i "runTCDF.py" --data data/covid-ger-extended-trim.csv --epochs=10000 --kernel_size=14 --hidden_layers=1 --learning_rate=0.001 --optimizer="Adam" --dilation_coefficient=14 --significance=0.8 --plot
```

Increasing the number of hidden layers again to two indicates that this trend only continues, leading to even higher link labels in the graph visualization.

```python
%run -i "runTCDF.py" --data data/covid-ger-extended-trim.csv --epochs=10000 --kernel_size=14 --hidden_layers=2 --learning_rate=0.001 --optimizer="Adam" --dilation_coefficient=14 --significance=0.8 --plot
```

## Italy

For the Italy data, the first experiment documented above is repeated. As can be observed, convergence is not as good as before with the hyper parameter settings used. Still, causal links are identified.

```python
%matplotlib inline
%run -i "runTCDF.py" --data data/covid-it-trim.csv --epochs=10000 --kernel_size=7 --hidden_layers=0 --learning_rate=0.001 --optimizer="Adam" --dilation_coefficient=7 --significance=0.8 --plot
```

We repeat the experiment with an increased learning rate of 0.001 to see if convergence can be improved while keeping the other hyper parameters - which is the case.

```python
%matplotlib inline
%run -i "runTCDF.py" --data data/covid-it-trim.csv --epochs=10000 --kernel_size=7 --hidden_layers=0 --learning_rate=0.01 --optimizer="Adam" --dilation_coefficient=7 --significance=0.8 --plot
```

And additinally, we also increase the kernel size as well as the dilation coefficient to 14. On this data set however, this does not lead to longer causal links, in contrast to the Germany data set as shown above.

```python
%matplotlib inline
%run -i "runTCDF.py" --data data/covid-it-trim.csv --epochs=10000 --kernel_size=14 --hidden_layers=0 --learning_rate=0.01 --optimizer="Adam" --dilation_coefficient=14 --significance=0.8 --plot
```

## United Stated of America

Finally, we run the first experiment applied to the Germany and Italy data sets also on the US Covid-19 data. However, we immediatly increase the learning rate to 0.01 as we did for italy, as the same variable columns are included. Again, we obtain possible causal links.

```python
%matplotlib inline
%run -i "runTCDF.py" --data data/covid-us-trim.csv --epochs=10000 --kernel_size=7 --hidden_layers=0 --learning_rate=0.01 --optimizer="Adam" --dilation_coefficient=7 --significance=0.8 --plot
```
